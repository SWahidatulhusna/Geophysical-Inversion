```python
xs=np.array([20,24,60,77,50])
ys=np.array([31,86,88,52,22])
t0=25
vp=4

#model awal
x0=40
y0=60

tcal = np.sqrt((xs - x0)**2 + (ys - y0)**2)/vp

# Penambahan  Noise
m, s = 0, 2
noise = np.random.normal(m, s, len(tcal))
tobs = tcal + noise
```

```python
iterasi= 50
x= random.sample(range(0, 500),iterasi) #Mc Sample untuk kedalaman
y= random.sample(range(0, 750),iterasi)
# X, Y = np.meshgrid(x, y, sparse=True)
ndata = len(x)
Terr = []
for i in range(len(x)):
    for j in range(len(y)):
        e_total = 0
        for k in range(len(xs)):
            ti = np.sqrt((xs[k] - x[i])**2 + (ys[k] - y[j])**2)/vp
            e = abs(ti - tobs[k])
            Erms = np.sqrt((1/(len(tobs)))*(e)**2)
            e_total += e
        Terr.append(e_total)
print("Erms",min(Terr))
Terr0 = np.array(Terr)
Terr1 = Terr0.reshape(ndata, ndata)
Terr1 = Terr1.transpose()

e_value = np.argwhere(Terr1==Terr1.min()).ravel()
xe_baru = x[e_value[1]]
ye_baru = y[e_value[0]]
print("Koordinat Episenter Baru:\n")
print(f'X = {xe_baru}\nY = {ye_baru}')

print(len(Terr))
```
